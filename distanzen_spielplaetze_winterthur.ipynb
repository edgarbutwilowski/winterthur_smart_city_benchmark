```python
import osmnx
import geopandas
import matplotlib
import contextily
```

Gemeindegrenzen von Winterthur holen:

```python
winterthur_grenze = osmnx.geocode_to_gdf("Winterthur, Switzerland")
winterthur_grenze.plot()
```

Spielplätze Winterthur holen:

```python
spielplaetze_in_winterthur = osmnx.features_from_place(query="Winterthur, Switzerland", tags={"leisure": "playground"})
spielplaetze_in_winterthur.head()
```

Kartenvorschau der Spielplätze in Winterthur:

```python
spielplaetze_in_winterthur_map = spielplaetze_in_winterthur.to_crs(epsg=3857)
spielplaetze_in_winterthur_plot = spielplaetze_in_winterthur_map.plot(alpha=0.5, edgecolor='k', color='red')
contextily.add_basemap(spielplaetze_in_winterthur_plot)
```

```python
winterthur_grenze_map = winterthur_grenze.to_crs(epsg=3857)
winterthur_grenze_plot = winterthur_grenze_map.plot(alpha=0.5, edgecolor='k')

spielplaetze_in_winterthur_map = spielplaetze_in_winterthur.to_crs(epsg=3857)
spielplaetze_in_winterthur_plot = spielplaetze_in_winterthur_map.plot(ax = winterthur_grenze_plot, alpha=0.5, edgecolor='k', color='red')

contextily.add_basemap(spielplaetze_in_winterthur_plot)
```

Alle Gebäude in Winterthur:

```python
gebaeude_in_winterthur = osmnx.features_from_place(query="Winterthur, Switzerland", tags={"building": True})
gebaeude_in_winterthur.plot()
```
